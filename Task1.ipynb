Data Import and Cleaning

Data Manipulation

```python
metrices = ["SystolicBP","DiastolicBP","HeartRate( bpm)"]
print(df[metrices].agg(["mean","min","max"]))
```

This code snippet computes the mean, minimum, and maximum values for the specified metrics in the DataFrame df.The .agg() method is used to apply aggregation functions ("mean", "min", "max") across the specified columns.

Computing BMI

```python
df["BMI"] = df["Weight(kg)"]/((df["Height(cm)"]/100)**2)
df
```

This code calculates the Body Mass Index (BMI) for each individual in the dataset and adds it as a new column in the DataFrame df.df["Height(cm)"]/100: Converts height from centimeters to meters.(df["Height(cm)"]/100)**2: Squares the height (in meters) as required by the BMI formula.df["Weight(kg)"]/((df["Height(cm)"]/100)**2): Divides the weight (in kilograms) by the squared height (in meters) to calculate BMI.The computed BMI values are assigned to a new column named "BMI" in the DataFrame df.

Basic Analysis

```python
abnormal_bp = df[(df["SystolicBP"]>140)|(df["DiastolicBP"]>90)]
abnormal_bp
```

This code filters the dataset to identify individuals with abnormal blood pressure readings, which are indicative of hypertension or high blood pressure.Systolic Blood Pressure (SystolicBP) exceeds 140 mmHg.Diastolic Blood Pressure (DiastolicBP) exceeds 90 mmHg.The | operator represents a logical OR, so individuals meeting either condition will be included.Rows where the above condition is True are selected from the DataFrame df.The filtered rows are assigned to a new DataFrame, abnormal_bp.Outputs the abnormal_bp DataFrame, showing individuals with high blood pressure.



```python
abnormal_hr = df[df["HeartRate( bpm)"]>100]
abnormal_hr
```

```python
def health_risk(row):
  if row["BMI"]<18.5 or row["BMI"]>30 or row["SystolicBP"]>140 or row["DiastolicBP"]>90:
    return "High Risk"
  elif 25<= row["BMI"]<=30 or 130<=row["SystolicBP"]<=140 or 80<=row["DiastolicBP"]<=90:
    return "Moderate Risk"
  else:
    return "Low Risk"
df["HealthRisk"]=df.apply(health_risk,axis =1)
df
```

This code snippet performs two primary tasks:
Identifies individuals with abnormal heart rates.
Assigns a health risk category to each individual based on multiple health indicators (BMI, blood pressure, etc.)

Visualization

```python
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-01", "2023-01-02", "2023-01-03"],
    "PatientID": [1, 1, 1, 2, 2, 2],
    "HeartRate": [72, 75, 70, 80, 85, 82],
    "SystolicBP": [120, 118, 115, 140, 138, 135],
    "DiastolicBP": [80, 78, 75, 90, 88, 85]
})

df["Date"] = pd.to_datetime(df["Date"])  # Convert 'Date' column to datetime
unique_patients = df["PatientID"].unique()  # Get unique Patient IDs

plt.figure(figsize=(12, 8))
for patient_id in unique_patients:
    patient_data = df[df["PatientID"] == patient_id]
    plt.plot(patient_data["Date"], patient_data["HeartRate"], marker="o", label=f"Heart Rate - Patient {patient_id}")
    plt.plot(patient_data["Date"], patient_data["SystolicBP"], marker="o", label=f"Systolic BP - Patient {patient_id}")
    plt.plot(patient_data["Date"], patient_data["DiastolicBP"], marker="o", label=f"Diastolic BP - Patient {patient_id}")

plt.xlabel("Date")
plt.ylabel("Values")
plt.title("Changes in Heart Rate and Blood Pressure for All Patients")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

This code generates a visualization of the changes in heart rate, systolic blood pressure (BP), and diastolic blood pressure (BP) over time for multiple patients.It displays trends in heart rate and blood pressure for individual patients over time.It allows side-by-side comparison of vital metrics between patients.It highlights how each metric fluctuates across the observed dates.



```python
avg_bmi = df.groupby('PatientID')['BMI'].mean()
avg_bmi.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.xlabel('PatientID')
plt.ylabel('Average BMI')
plt.title('Average BMI of Patients')
plt.show()
```

This code calculates and visualizes the average Body Mass Index (BMI) for each patient in the dataset.
Groups the dataset by the PatientID column.
Calculates the mean BMI for each group (i.e., each patient).
Stores the result in avg_bmi, a pandas Series with PatientID as the index and average BMI values as the data.
Creates a bar plot of the average BMI values for each patient:
kind='bar': Specifies a bar chart.
figsize=(10, 6): Sets the size of the plot to make it visually clear.
color='skyblue': Assigns a light blue color to the bars for aesthetic appeal.
Adds labels for the x-axis (PatientID) and y-axis (Average BMI).
Titles the plot 'Average BMI of Patients' for context.
Uses plt.show() to display the final visualization.


```python
risk_distribution = df['HealthRisk'].value_counts()
risk_distribution.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=140)
plt.title('Health Risk Group Distribution')
plt.ylabel('')
plt.show()
```

This code generates a pie chart that illustrates the distribution of individuals across different health risk categories.
The value_counts() method counts the occurrences of each unique value in the "HealthRisk" column.
The result, risk_distribution, is a pandas Series where:
The index represents the health risk categories ("High Risk", "Moderate Risk", "Low Risk").
The values represent the counts of individuals in each category.
kind='pie': Specifies a pie chart.
autopct='%1.1f%%': Displays the percentage of each category in the pie slices, formatted to one decimal place.
figsize=(8, 8): Sets the size of the figure for better visibility.
startangle=140: Rotates the chart to start at 140 for better alignment.
Adds a title to describe the chart: 'Health Risk Group Distribution'.
Removes the y-axis label (irrelevant for a pie chart).
Renders and displays the pie chart.
